## Transforming Data Using a Function or Mapping
### For many datasets, you may wish to perform some transformation based on the values in an array, Series, or column in a DataFrame. Consider the following data collected about various kinds of meat:

````python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
````

````python
data['food'].str.lower()
````

````python
lowercased = data['food'].str.lower()
lowercased
````

### Step 3: Apply the map method on a Series by accepting a function or dict-like object containing a mapping

````python
data['animal'] = data['food'].str.lower().map(meat_to_animal)
````

````python
data
````

# Replacing Values
Filling in missing data with the fillna method is a special case of more general value
replacement. As you’ve already seen, map can be used to modify a subset of values in
an object.
## `replace` provides a simpler and more flexible way to do so. Let’s consider this Series:

````python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
````

The -999 values might be sentinel values for missing data. To replace these with NA
values that pandas understands, we can use replace, producing a new Series (unless
you pass inplace=True):

````python
data1 = data.replace([-999, -1000], np.nan)
````

````python
data1.fillna(data1.mean())
````

If you want to replace multiple values at once, you instead pass a list and then the
substitute value:

````python
data.replace([-999, -1000], np.nan)
````

To use a different replacement for each value, pass a list of substitutes:

````python
data.replace([-999, -1000], [np.nan, 0])
````

The argument passed can also be a dict:

````python
data.replace({-999: np.nan, -1000: 0})
````

# Discretization and Binning
Continuous data is often discretized or otherwise separated into “bins” for analysis.
Suppose you have data about a group of people in a study, and you want to group
them into discrete age buckets:

````python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
````

### Let’s divide these into bins of 18 to 25, 26 to 35, 36 to 60, and finally 61 and older. To do so, you have to use cut, a function in pandas:

````python
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
````

````python
cats
````

### The object pandas returns is a special Categorical object. The output you see describes the bins computed by pandas.cut. You can treat it like an array of strings indicating the bin name; internally it contains a categories array specifying the distinct category names along with a labeling for the ages data in the codes attribute:

````python
cats.codes
````

````python
cats
````

````python
cats.categories
````

````python
pd.value_counts(cats)
````

### Note that pd.value_counts(cats) are the bin counts for the result of pandas.cut.
### Consistent with mathematical notation for intervals, a parenthesis means that the side is open, while the square bracket means it is closed (inclusive). You can change which side is closed by passing right=False:

````python
pd.cut(ages, [18, 26, 36, 61, 100], right=False)
````

### You can also pass your own bin names by passing a list or array to the labels option:

````python
ages
````

````python
bins = [18, 35, 70, 100]
bins
````

````python
pd.cut(ages, bins)
````

````python
group_names = ['Youth', 'MiddleAged', 'Senior']
group_names
````

````python
cats = pd.cut(ages, bins, labels=group_names)
````

````python
pd.value_counts(cats)
````

### If you pass an integer number of bins to cut instead of explicit bin edges, it will compute equal-length bins based on the minimum and maximum values in the data. Consider the case of some uniformly distributed data chopped into fourths:

````python
data = np.random.rand(20)
data
````

````python
cats = pd.cut(data, 4, precision=2)
````

````python
cats
````

````python
cats.value_counts()
````

### The precision=2 option limits the decimal precision to two digits.
### A closely related function, qcut, bins the data based on sample quantiles. Depending on the distribution of the data, using cut will not usually result in each bin having the same number of data points. Since qcut uses sample quantiles instead, by definition you will obtain roughly equal-size bins:

````python
data = np.random.randn(1000)  # Normally distributed
````

````python
cats = pd.qcut(data, 4, precision=2)  # Cut into quartiles
````

````python
cats
````

````python
pd.value_counts(cats)
````

### Similar to cut you can pass your own quantiles (numbers between 0 and 1, inclusive):

````python
cats =pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])
cats
````

````python
cats.value_counts()
````
